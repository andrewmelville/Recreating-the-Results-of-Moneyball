# Recreating the Results of Moneyball

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Introduction

The Data contains 21 performance statistics for 322 baseball players. The objective here is to broadly recreate the success of the Oakland A's, subject of Michael Lewis' "Moneyball", and build some models that predict a players salary based their batting performance statistics. Residuals of this model should then highlight batters that are currently under or over paid for their abilities and so can be identified as transfer targets. 

As batting and pitching are very distinct skills, we limit this analysis to batting performance.

## Preprocessing

```python
hitters = pd.read_csv("Hitters.csv")
```

```python
print('Number of Batters in Dataset:', hitters.shape[0])
```

Unlike the rest of the features, both the 'League' and 'Division' features stand out as being strings and not numercial values. However there is no discernible skill difference between leagues so this can be ignored, and the 'Division' feature can be represented as two presence features in the event that we wish to compare differences between divisions.

### Division/League

```python
hitters['West'] = [1 if div == 'W' else 0 for div in hitters['Division']]
hitters['East'] = [1 if div == 'E' else 0 for div in hitters['Division']]
hitters.drop(['Division', 'League', 'NewLeague'], axis=1, inplace=True)
```

### Hitters or Pitchers?
In line with the stated aim of the model, we isolate the data to performance over the preceeding season, as analysis of career statistics may yeild fading stars, while performance in the previous season is the best indicator of performance in the upcoming season. 

```python
hitters.drop(['PutOuts', 'Assists', 'Errors'], axis=1, inplace=True)
hitters.columns
```

### Previous Season Stats
As stated above, we limit analysis to performance over the previous season, and also disregard the number of years a player has been in the league.

```python
hitters.drop([feat for feat in hitters.columns if feat[0] == 'C'], axis=1, inplace=True)
# hitters.drop(['Years'], axis=1, inplace=True)
hitters.columns
```

### Free Agents
We now deal with the players with missing data for their 'Salary' feature. This may be due to missing data, but far more likely is that they are out of a contract at the time of data collection. As a result, filling this value with the mean of the data would only serve to obfuscate the results of any model, and a choice of setting these to $0 per year (which is closest to reality) would still not yeild undervalued players, as they may still attract interest from wealthy clubs who are prepared to pay them a sizeable wage. Short of collecting data on their salary for the previous season (to be used as expected salary for the next season) we are forced to eliminate these players from our analysis.

```python
# Remving players with no salary data
hitters = hitters[np.isnan(hitters['Salary'].values) == False]
```

```python
hitters.head()
```

### Pairpplotting Remaining Data and OnBase %

```python
sns.pairplot(hitters)
```

The pairplot shows clear correlation between hits and times at bat. The reason for this is obvious: you will make more hits if you attempt more hits. Numbers of runs and homeruns are similarly correlated. This prompts some dimensionality reduction in the form of variable creation. 

Here we create a variable which captures a players success rate when at bat, "On-base percentage". This is defined as the number of times a player gets to first base by any means, including both hits and walks. The assumption that this choice makes is that the number of times a player gets on base is correlated to their runs (and so success as a player). As Peter Brand says, "Your goal shouldn't be to buy players, your goal should be to buy wins. And in order to buy wins, you need to buy runs."

First, we create the On-Base % variable. Then we validate this assumption of correlation. Fiinally, we proceed with this being our primary predictor of salary.

```python
# Creating OB%
times_on_base = hitters['Hits'] + hitters['Walks']
hitters['OB %'] = times_on_base / hitters['AtBat'].values

# Creating Total Runs column
hitters['Total Runs'] = hitters['Runs'] + hitters['HmRun']

# Creating RUN%
hitters['Run%'] = hitters['Total Runs'] / hitters['AtBat']
```

```python
sns.pairplot(hitters[['OB %', 'Total Runs', 'Run%']])
```

The pairplot shows clear correlation between getting on base and both number of runs and runs as a percentage of your times at bat.

## Train/Test Split

Here we split our full data into a training set to train our models on, and a test set to validate the results of the models on. As always, the model is blind to the test set until the moment of prediction.

```python
# Label Name column
hitters.rename(columns={'Unnamed: 0': 'Name'}, inplace=True)

# Get number of observations
n = hitters.shape[0]

# Split into 80% for training, 20% for testing
hitters_train = hitters.loc[:n*0.8]
hitters_test = hitters.loc[0.8*n:]
```

## Models
### Linear Regression
We first attempt to regress a players last known salary onto the players on bat percentage including a y intercept. We do not standardise these quantities in order to preserve model interpretibility.

The choice of a linear model implicitly makes the assumption that for every OB% point, a player will earn a constant amount per year. This certainly captures our goal of buying runs, but is not necessarily the most pertinent choice for player valuation as it is commonly observed that better players are over-rated by a larger amount. To be truthful to both the data and our goal or identifying significantly over-rated players to release and under-rated players to sign, we may want to account for this non-constant variance and model a non-constant mean response.

Later we will fit a GLM which allows for more flexibility in fitting mean responses to better capture this phenomenon.

```python
# Limiting data to variables considered for the linear regression
hitters_train_lr = hitters_train[['Name', 'Salary', 'OB %', 'West', 'East']]
hitters_test_lr = hitters_test[['Name', 'Salary', 'OB %', 'West', 'East']]
```

```python
from sklearn.linear_model import LinearRegression

# Calling and fitting a linear regression with an intercept term
salary_lr = LinearRegression(fit_intercept=False)
salary_lr.fit(hitters_train_lr[['OB %', 'West', 'East']].values, hitters_train_lr['Salary'].values.reshape(-1, 1))
```

```python
# Evaluation Model Coefficients
print(f"OB% Coefficient = {salary_lr.coef_}")
print(f"Intercept = {salary_lr.intercept_}")
```

### Interpretation of coefficients
The linear regression model fits two lines (East Div and West Div) with the constraint that they both compensate each extra OB% point by the same amount.

The fit predicts that each extra OB% will earn a player \\$30,000 a year on average in both leagues, but that every is compensated $117,000 less in the East Div than the West Div for the same performance.

```python
# Evaluate model performance
print(f" R^2 = {salary_lr.score(hitters_train_lr[['OB %', 'West', 'East']].values, hitters_train_lr['Salary'].values.reshape(-1, 1))}")
```

While the low $R^2$ score initially looks like a badly fitted model, this is actually useful to us as it both valiates a positive correlation in the variables, and leaves room for us to capitalise on the residuals that correspond to what the model thinks are over- or under-valued players.

```python
# Plot true and fitted values
plt.figure(figsize=(15,7))

plt.title('True and Fitted Values in the On-Base % - Salary Model')
plt.xlabel('On-Base %')
plt.ylabel('Salary ($1000)')
plt.xlim((0.25,0.51))

plt.scatter(hitters_train_lr['OB %'][hitters_train_lr['East'] == 1], hitters_train_lr['Salary'][hitters_train_lr['East'] == 1], s=3)
plt.scatter(hitters_train_lr['OB %'][hitters_train_lr['East'] == 0], hitters_train_lr['Salary'][hitters_train_lr['East'] == 0], s=3)

plt.plot([0.52*i/100 for i in range(40,100)], salary_lr.predict([[0.52*i/100,0,1] for i in range(40,100)]), label='East Predicted Salary', color='b')
plt.plot([0.52*i/100 for i in range(40,100)], salary_lr.predict([[0.52*i/100,1,0] for i in range(40,100)]), label='West Predicted Salary', color='orange')

# plt.plot(hitters_train_lr['OB %'][hitters_train_lr['West'] == 0], salary_lr.predict(hitters_train_lr[['OB %', 'West', 'East']][hitters_train_lr['West'] == 0].values), label='East Predicted Salary', s=3)

plt.legend()
```

The plot above shows a players true salary against the models prediction. By including two levels (East and West division) we have effectively fit two models with the constraint that the gradient of the line remains constant. This is a valid assumption, as we do not expect there to be significant difference in the amount that extra skill is compensated between 
coasts. We would however expect to see a difference in the amount that the same skill is compensated between the two divisions, and this is refelcted in the different estimated y-intercepts. The model predicts that a player with any given OB% can expect to earn $117,000 per year more in the East Division on average.

## Generalised Linear Model
Here we fit two GLMs on the same training data as the Linear Model, using the same covariates.

### Gamma with Log Link Function
The Gamma distribution is a flexible choice for the data we have collected, being strictly positive and with distribution shape close to that of a Gamma. However we have fit a GLM with the log link function. 

```python
import statsmodels.api as sm

salary_log = sm.GLM(hitters_train_lr['Salary'].values, hitters_train_lr[['OB %', 'East', 'West']].values,
                    family=sm.families.Gamma(sm.families.links.log))

salary_log_fit = salary_log.fit()
salary_log_fit.summary()
```

```python
plt.figure(figsize=(15,7.5))

plt.title('Gamma with Log Link Function Fitted Values')
plt.xlabel('On-Base %')
plt.xlim((0.25,0.51))
plt.ylabel('Salary ($1000)')


plt.plot([0.52*i/100 for i in range(40,100)],salary_log_fit.predict([[0.52*i/100,0,1] for i in range(40,100)]), 'orange', label='West Fitted Curve')
plt.plot([0.52*i/100 for i in range(40,100)],salary_log_fit.predict([[0.52*i/100,1,0] for i in range(40,100)]), 'b', label='East Fitted Curve')
plt.scatter(hitters_train_lr['OB %'][hitters_train_lr['East'] == 1], hitters_train_lr['Salary'][hitters_train_lr['East'] == 1], s=5, color='b')
plt.scatter(hitters_train_lr['OB %'][hitters_train_lr['East'] == 0], hitters_train_lr['Salary'][hitters_train_lr['East'] == 0], s=5, color='orange')

plt.legend()
```

The curve shows a good sweep close to data points, with higher variance at larger values. This corresponds with our the intuition we have about good but overrated players.

### Gamma with Canonical Link Function
The canonical link function may yeild a curve closer to the data, which we test and quantify here.

```python
import statsmodels.api as sm

salary_canon = sm.GLM(hitters_train_lr['Salary'].values, hitters_train_lr[['OB %', 'East', 'West']].values,
                    family=sm.families.Gamma())

salary_canon_fit = salary_canon.fit()
salary_canon_fit.summary()
```

```python
salary_canon_fit.predict([[0.52*i/100,0,1] for i in range(50,100)])
salary_canon_fit.predict([[0.52*i/100,1,0] for i in range(50,100)])
```

```python
plt.figure(figsize=(15,7.5))
plt.title('Gamma with Canonical Link Function Fitted Values')
plt.xlabel('On-Base %')
plt.xlim((0.25,0.51))
plt.ylabel('Salary ($1000)')

# Fitted lines
plt.plot([0.52*i/100 for i in range(40,100)], salary_canon_fit.predict([[0.52*i/100,0,1] for i in range(40,100)]), 'orange', label='West Fitted Curve')
plt.plot([0.52*i/100 for i in range(40,100)], salary_canon_fit.predict([[0.52*i/100,1,0] for i in range(40,100)]), 'b', label='East Fitted Curve')

# Data
plt.scatter(hitters_train_lr['OB %'][hitters_train_lr['East'] == 1], hitters_train_lr['Salary'][hitters_train_lr['East'] == 1], s=5, color='b')
plt.scatter(hitters_train_lr['OB %'][hitters_train_lr['East'] == 0], hitters_train_lr['Salary'][hitters_train_lr['East'] == 0], s=5, color='orange')

plt.legend()
```

The canonical link function appears to over fit to the salary of players with the highest On-Base percentages. As both models have the same number of parameters, we can compare their respective performances by evaluating their deviance statistics. This is a measure of the model's residual sum of squares compared to that of the fully saturated model.

```python
#RSS of canonical and logistic GLMs
print(f"Canonical link Deviance: {salary_canon_fit.deviance}")
print(f"Logistic link Deviance: {salary_log_fit.deviance}")
```

Both models have very similar deviances, achieve similar minimisations of their respective likelihoods, and hold very similar AIC scores. To differentiate between them we must evaluate their fit to the data left aside for testing.

## Model Evaluation
Here we evaluate and compare the linear model with the two GLMs. From their performance on the training data, we expect the GLMs to out perform the linear model, and hope that the test data is enough to significantly determine the superior GLM.

```python
plt.figure(figsize=(15,7.5))

plt.title('Gamma with Log Link Function Test Performance')
plt.xlabel('On-Base %')
plt.xlim((0.2,0.56))
plt.ylabel('Salary ($1000)')
plt.ylim((0,2500))

plt.plot([0.57*i/100 for i in range(20,100)],salary_lr.predict([[0.57*i/100,0,1] for i in range(20,100)]), 'orange', label='West Fitted Curve')
plt.plot([0.57*i/100 for i in range(20,100)],salary_lr.predict([[0.57*i/100,1,0] for i in range(20,100)]), 'b', label='East Fitted Curve')

plt.scatter(hitters_test_lr['OB %'][hitters_test_lr['East'] == 1], hitters_test_lr['Salary'][hitters_test_lr['East'] == 1], s=5, color='b')
plt.scatter(hitters_test_lr['OB %'][hitters_test_lr['East'] == 0], hitters_test_lr['Salary'][hitters_test_lr['East'] == 0], s=5, color='orange')

plt.legend()
```

```python
plt.figure(figsize=(15,7.5))

plt.title('Gamma with Log Link Function Test Performance')
plt.xlabel('On-Base %')
plt.xlim((0.2,0.56))
plt.ylabel('Salary ($1000)')

plt.plot([0.57*i/100 for i in range(20,100)],salary_log_fit.predict([[0.57*i/100,0,1] for i in range(20,100)]), 'orange', label='West Fitted Curve')
plt.plot([0.57*i/100 for i in range(20,100)],salary_log_fit.predict([[0.57*i/100,1,0] for i in range(20,100)]), 'b', label='East Fitted Curve')

plt.scatter(hitters_test_lr['OB %'][hitters_test_lr['East'] == 1], hitters_test_lr['Salary'][hitters_test_lr['East'] == 1], s=5, color='b')
plt.scatter(hitters_test_lr['OB %'][hitters_test_lr['East'] == 0], hitters_test_lr['Salary'][hitters_test_lr['East'] == 0], s=5, color='orange')

plt.legend()
```

```python
plt.figure(figsize=(15,7.5))

plt.title('Gamma with Canoncial Link Function Test Performance')
plt.xlabel('On-Base %')
plt.xlim((0.2,0.56))
plt.ylabel('Salary ($1000)')
plt.ylim((0,2500))

plt.plot([0.57*i/100 for i in range(20,100)],salary_canon_fit.predict([[0.57*i/100,0,1] for i in range(20,100)]), 'orange', label='West Fitted Curve')
plt.plot([0.57*i/100 for i in range(20,100)],salary_canon_fit.predict([[0.57*i/100,1,0] for i in range(20,100)]), 'b', label='East Fitted Curve')

plt.scatter(hitters_test_lr['OB %'][hitters_test_lr['East'] == 1], hitters_test_lr['Salary'][hitters_test_lr['East'] == 1], s=5, color='b')
plt.scatter(hitters_test_lr['OB %'][hitters_test_lr['East'] == 0], hitters_test_lr['Salary'][hitters_test_lr['East'] == 0], s=5, color='orange')

plt.legend()
```

As expected, the linear model is able to recover the mean of the test data and little else. The natural non-linearity of salary is too much for the linear model to handle outside of a small region near the mode of the data.

The Logistic-Gamma model passes a quick inspection of its performance of the test data. There is a clear passage through the mean of the point cloud, and room for inference on the residuals in the hopes of picking up a productive player for cheap.

While the Canonical-Model does fit all the test data within the range of the original training data well, the choice of link function ordains that their must be a singularity somewhere in the parameter space of the linear predictor. Despite (somewhat miraculously) passing through our most extreme data point, this model serves best as a reminder that we must not extrapolate beyond the relationships learned from the range of the training data. As such, the model remains valid for values below this point (around 0.54).

We now quantify the performance of these models on the test data with some performance statistics.

### RSS on Training and Test Data

```python
# Removing outlying residual
hitters_test_lr_wo_res = hitters_test_lr[hitters_test_lr['OB %'] < 0.53]

# Calculating MSE on testing data without residual
np.sum((hitters_test_lr['Salary'].values - salary_log_fit.predict(hitters_test_lr[['OB %', 'East', 'West']])) ** 2) / hitters_test_lr.shape[0]
```

```python
# Calculating MSE on training data
MSE_train = {}
MSE_train['Linear'] = np.sum((hitters_train_lr['Salary'].values - salary_lr.predict(hitters_train_lr[['OB %', 'East', 'West']])) ** 2) / hitters_train_lr.shape[0]
MSE_train['Logistic'] = np.sum((hitters_train_lr['Salary'].values - salary_log_fit.predict(hitters_train_lr[['OB %', 'East', 'West']])) ** 2) / hitters_train_lr.shape[0]
MSE_train['Canonical'] = np.sum((hitters_train_lr['Salary'].values - salary_canon_fit.predict(hitters_train_lr[['OB %', 'East', 'West']])) ** 2) / hitters_train_lr.shape[0]

# Calculating MSE on testing data with residual
MSE_test = {}
MSE_test['Logistic'] = np.sum((hitters_test_lr['Salary'].values - salary_log_fit.predict(hitters_test_lr[['OB %', 'East', 'West']])) ** 2) / hitters_test_lr.shape[0]
MSE_test['Canonical'] = np.sum((hitters_test_lr['Salary'].values - salary_canon_fit.predict(hitters_test_lr[['OB %', 'East', 'West']])) ** 2) / hitters_test_lr.shape[0]

# Removing outlying residual
hitters_test_lr_wo_res = hitters_test_lr[hitters_test_lr['OB %'] < 0.53]

# Calculating MSE on testing data without residual
MSE_test_fix = {}
MSE_test_fix['Logistic'] = np.sum((hitters_test_lr_wo_res['Salary'].values - salary_log_fit.predict(hitters_test_lr_wo_res[['OB %', 'East', 'West']])) ** 2) / hitters_test_lr_wo_res.shape[0]
MSE_test_fix['Canonical'] = np.sum((hitters_test_lr_wo_res['Salary'].values - salary_canon_fit.predict(hitters_test_lr_wo_res[['OB %', 'East', 'West']])) ** 2) / hitters_test_lr_wo_res.shape[0]
```

```python
MSE_test_fix['Canonical']
```

```python
## Visualising MSE of three candidate models.

fig, axs = plt.subplots(1,3,figsize=(15,5))

# Plotting barplots and setting titles
axs[0].bar([model for model in MSE_train.keys()], [MSE_train[model] for model in MSE_train.keys()])
axs[0].title.set_text('Training MSE')

axs[1].bar([model for model in MSE_test.keys()], [MSE_test[model] for model in MSE_test.keys()])
axs[1].title.set_text('Testing MSE')

axs[2].bar([model for model in MSE_test_fix.keys()], [MSE_test_fix[model] for model in MSE_test_fix.keys()])
axs[2].title.set_text('Testing MSE with Removed Residual')
```

On the training data it is clear that the GLMs outperform the linear model by orders of magnitude. However, they diverge drastically when tasked with evaluating new players. The second plot above shows that the Canonical model performs much worse, and some digging shows that this is due its poor performance on one player with an OB % of 0.55. This is outside the range of our training data, and results in an enormous residual which tanks the Canonical model's MSE. Without this extreme value, the GLMs are indistinguishable. 

Such a high OB% is rare, but we would still like to be able to evaluate players of this calibre, and so on this basis we choose the Logistic model with Gamma-distributed errors as our final choice of model.

## Chosen Model Interpretation
Our chosen logistic link function equates a linear increase in On-Base % to an exponential increase in expected salary. By choosing to encode the East and West divisions as seperate columns, we have contrained the model to fit two curves with the same response to OB%, but allowed flexibility in the absolute levels of these curves. This means that the model predicts that while both coasts compensate extra skill equally, the East coast team pay on average 1.22 times that of the West coast teams for a player with the same OB%.

The model advises that baseball Front Office staff should scout and attract players who have large negative residuals, who are earmarked as heavily underrated, and avoid at all costs those players above the line who are paid extortionate amounts for middling performances.

The equivalent advice for baseball players is to move to Philadelphia, not Fresno.

# Comments
Despite good performance in the final chosen model, we would do well to include the number of years a player has played in the league as a covariate. This is because we would expect older players to demand larger contracts on average, which is an assumption validated by the pairplots plotting in the first section.
